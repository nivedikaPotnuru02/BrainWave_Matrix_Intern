```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display charts in the notebook
%matplotlib inline
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
#Total Sales by Category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', title='Sales by Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Sales Over Time
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)
df.resample('M')['Sales'].sum().plot(title='Monthly Sales Trend')
plt.ylabel('Sales')
plt.show()
```

```python
#Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', title='Top 10 Products by Sales')
plt.ylabel('Sales')
plt.show()
```

```python
#Sales by Region
region_sales = df.groupby('Region')['Sales'].sum()
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title('Sales by Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Profit vs Discount
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title('Profit vs Discount')
plt.show()
```

```python
#Total Sales and Profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
# Sales by Category and Sub-Category
# Sales by Category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("Sales by Category:")
print(category_sales)

# Sales by Sub-Category
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
print("\nSales by Sub-Category:")
print(subcat_sales)
```

```python
#Sales by Category and Sub-Category Visualize it with Bar Charts
# Bar Chart for Category
category_sales.plot(kind='bar', title='Sales by Category', ylabel='Total Sales', xlabel='Category')
plt.xticks(rotation=0)
plt.show()

# Bar Chart for Sub-Category
subcat_sales.plot(kind='bar', figsize=(10, 5), title='Sales by Sub-Category', ylabel='Total Sales', xlabel='Sub-Category', color='skyblue')
plt.xticks(rotation=45)
plt.show()
```

```python
#Loss-Making Categories (Profit < 0)
# Total profit by Sub-Category
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()

# Filter loss-making only
loss_making = subcat_profit[subcat_profit < 0]

print("Loss-Making Sub-Categories:")
print(loss_making)
```

```python
#Plot Loss-Making Sub-Categories
loss_making.plot(kind='bar', color='red', title='Loss-Making Sub-Categories')
plt.ylabel('Total Profit')
plt.xlabel('Sub-Category')
plt.xticks(rotation=45)
plt.show()
```
